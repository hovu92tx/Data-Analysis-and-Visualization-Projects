```python
import pandas as pd
import plotly.graph_objs as go
import plotly.io as pio
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.offline as py
```

```python
#import datasets 1
dataset1 = pd.read_csv('covid.csv')
# Set display option to show all columns
pd.set_option('display.max_colum', None)
print(dataset1.head(10))
print(dataset1.describe())
dataset1.info()

# Drop NewCases, NewDeaths, NewRecovered columns because of missing values. 
dataset1.drop(['NewCases', 'NewDeaths', 'NewRecovered'],  axis=1, inplace=True)

#plot the top 15 countries base on total cases.
plot1= px.bar(dataset1.head(15), x= 'Country/Region', y= 'TotalCases', color='TotalCases', height= 500, hover_data='Country/Region')
#plot1.show()
plot1.write_image('./Graphs/plot1.png')

#plot the top 15 countries base on total deaths and corlor by total deaths 
plot2= px.bar(dataset1.head(15), x= 'Country/Region', y= 'TotalCases', color='TotalDeaths', height=500, hover_data='Country/Region')
#plot2.show()
plot2.write_image('./Graphs/plot2.png')

# plot the top 15 countries base on total recovered and color by total recovered
plot3 = px.bar(dataset1.head(15), x= 'Country/Region', y= 'TotalCases', color='TotalRecovered', height=500, hover_data='Country/Region') 
plot3.write_image('./Graphs/plot3.png')


# #plot the Continents base on total cases.
# plot4= px.bar(dataset1.head(15), x= 'Continent', y= 'TotalCases', color='TotalCases', height= 500, hover_data='Continent')
# #plot1.show()
# plot4.write_image('./Graphs/plot4.png')

# #plot the Continents base on total deaths and corlor by total deaths 
# plot5= px.bar(dataset1.head(15), x= 'Continent', y= 'TotalCases', color='TotalDeaths', height=500, hover_data='Continent')
# #plot2.show()
# plot5.write_image('./Graphs/plot5.png')

# # plot the Continents base on total recovered and color by total recovered
# plot6 = px.bar(dataset1.head(15), x= 'Continent', y= 'TotalCases', color='TotalRecovered', height=500, hover_data='Continent') 
# plot6.write_image('./Graphs/plot6.png')

# Data Visualization through Bubble Charts-Continent Wise
plot7 = px.scatter(dataset1, x='Continent',y='TotalCases',  
           hover_data=['Country/Region', 'Continent'],  
           color='TotalCases', size='TotalCases', size_max=80)
plot7.write_image('./Graphs/plot7.png')

# Data Visualization through Bubble Charts by Total Test
plot8= px.scatter(dataset1.head(54), x='Continent',y='TotalTests', 
		hover_data=['Country/Region', 'Continent'], 
		color='TotalTests', size='TotalTests', size_max=80)
plot8.write_image('./Graphs/plot8.png')
```

```python
#import datasets 2
dataset2 = pd.read_csv('covid_grouped.csv')
# Set display option to show all columns
pd.set_option('display.max_colum', None)
print(dataset2.head(10))
print(dataset2.describe())
dataset2.info()

# Extract US data from the dataset 
df_US= dataset2.loc[dataset2["Country/Region"]=="US"]

# Plot US Confirmed Cases by Date
plot9 = px.bar(df_US, x="Date", y="Confirmed", color="Confirmed", height=400)
plot9.write_image('./Graphs/plot9.png')

# Plot US Recovered Cases by Date
plot10 = px.line(df_US,x="Date", y="Recovered", height=400)
plot10.update_layout(title = {'text':' US Recorvered Char by Date','x': 0.5, 'xanchor':'center'})
plot10.write_image('Graphs/plot10.png')

# Plot US Deaths by Date
plot11 = px.line(df_US,x="Date", y="Deaths", height=400)
plot11.update_layout(title= {
    'text': 'US\'s Deaths by Date',
    'x': 0.5,
    'xanchor':'center'
})
plot11.write_image('./Graphs/plot11.png')

# Plot US new Cases by Date

plot12= px.line(df_US,x="Date", y="New cases", height=400)
plot12.update_layout(title={
    'text': 'US New Cases by Date',
    'x': 0.5,
    'xanchor': 'center'
})
plot12.write_image('./Graphs/plot12.png')
```

```python
# Visualization of Data in terms of Maps
plot13 = px.choropleth(dataset2, 
			locations="iso_alpha", 
			color="Confirmed", 
			hover_name="Country/Region", 
			color_continuous_scale="Blues", 
			animation_frame="Date")
plot13.show()
plot13.write_html('./Graphs/plot13.html')
```
